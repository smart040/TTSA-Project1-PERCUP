```python
# now I want to get it in 3-dim 

def generate_circular_data(num_samples, radius):
    angles = np.linspace(0, 2*np.pi, num_samples)
    x = radius * np.cos(angles)
    y = radius * np.sin(angles)
    return x, y


# Parameters
num_samples = 1000
radius = 1


# Generate circular data
x, y = generate_circular_data(num_samples, radius)
z = np.zeros_like(x)  # Set z-coordinate to 0 for all points

circle_1 = [x-2,y,z]
circle_2 = [x+2,y,z]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(x-2, y, z)
ax.scatter(x+2, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Circular Data Set in 3D')

plt.show()
```

```python
#now I want to get the sphere


def generate_spherical_data(num_samples, radius):
    phi = np.random.uniform(0, np.pi, num_samples)  # Azimuthal angle
    theta = np.random.uniform(0, 2*np.pi, num_samples)  # Polar angle

    x = radius * np.sin(phi) * np.cos(theta)
    y = radius * np.sin(phi) * np.sin(theta)
    z = radius * np.cos(phi)

    return x, y, z

# Parameters
num_samples = 1000
radius = 1

# Generate spherical data
x_s, y_s, z_s = generate_spherical_data(num_samples, radius)
sphere_data = [x_s, y_s, z_s]

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(x_s, y_s, z_s)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Spherical Data Set')

plt.show()
```
